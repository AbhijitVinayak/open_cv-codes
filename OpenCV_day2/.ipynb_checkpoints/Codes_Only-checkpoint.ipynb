## Image handling in OpenCV
#### Opening an image

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread("Car.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Car image")
plt.axis("off")
plt.show()

print("The dimensions of the image are {dims}".format(dims=img.shape))
```

## Using a webcam


```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
cam = cv.VideoCapture(0)
cam.open(0)
while cam.isOpened():
    ret, frame = cam.read()
    if not ret:
       continue
    cv.imshow('Live feed', frame)
    key = cv.waitKey(1) & 0xFF
    if key == ord('q'):
       break
print("Done")
cv.destroyWindow('Live feed')
cv.destroyAllWindows()
cam.release()
```

## Video Handling using OpenCV

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
file = "CAT.mp4"
tdelay = 20
vid_reader = cv.VideoCapture(file)
while (vid_reader.isOpened()):
    ret, img = vid_reader.read()
    if not ret:
        break
    cv.imshow('Frame', img)
    key = cv.waitKey(tdelay) & 0xFF
    if key == ord('q'):
        break
cv.destroyAllWindows()
vid_reader.release()
```

### Video Recording using OpenCV

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
cam_no = 0
dest_file_name = 'RecordedFile.avi'
fps = 20.0
frame_size = (640, 480)
cam = cv.VideoCapture(cam_no)
rec_fourcc = cv.VideoWriter_fourcc(*'XVID')
rec = cv.VideoWriter(dest_file_name, rec_fourcc, fps, frame_size)
fno = 0
while cam.isOpened():
    ret, frame = cam.read()
    if not ret: 
        break
    cv.imshow("Live Feed", frame)
    rec.write(frame)
    fno += 1
    key = cv.waitKey(1) & 0xff
    if key == ord('q'):
        break
cam.release()
rec.release()
cv.destroyAllWindows()
print("{num_frames} frames written to file \"{fname}\"".format(num_frames=fno, fname=dest_file_name))
```
